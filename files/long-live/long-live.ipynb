```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import seaborn as sns

# visualization
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/home/otavio/ml/datasets/long-live/long-live.csv')
```

```python
data = data[['w', 'Years gained / lost', 'strength of science?']]
```

```python
data.rename(columns={'w': 'label', 'strength of science?': 'strength', 'Years gained / lost': 'years'}, inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
def convert(text):
    try:
        float(text)
        return 1
    except:
        return 0
```

```python
data = data[data['years'].apply(convert) == 1]
data['years'] = data['years'].astype(float)
```

```python
strength = {'suggestive': 1, 'good': 2, 'strong': 3}
data['strength'] = data['strength'].map(strength)
```

```python
from googletrans import Translator
gt = Translator()
```

```python
data['titulo'] = data['label'].apply(lambda x: gt.translate(text=x, src='en', dest='pt').text)
```
