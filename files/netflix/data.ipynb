```python
data = pd.read_csv('/home/otavio/ml/datasets/netflix/netflix_titles.csv')
data.info()
```

```python
data['added'] = data['date_added'].str.split(',', expand = True)[1]
data['release_year'] = data['release_year'].astype(int)
data['gap'] = data['added'].astype(int) - data['release_year']
```

```python
movies = data[data['type'] == 'Movie'].copy()
shows = data[data['type'] == 'TV Show'].copy()
```
