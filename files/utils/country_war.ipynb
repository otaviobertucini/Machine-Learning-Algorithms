```python
# data analysis and wrangling
import pandas as pd
import numpy as np
```

```python
# https://www.kaggle.com/fernandol/countries-of-the-world#countries%20of%20the%20world.csv
world = pd.read_csv('/home/otavio/ml/datasets/utils/guerra/countries of the world.csv')
```

```python
world
```

```python
world['Country'] = world['Country'].apply(lambda x: x.replace(' ', ''))
```

```python
hdi = pd.read_csv('/home/otavio/ml/datasets/utils/guerra/hdi.csv')
```

```python
hdi = hdi[['Country', '2018']]
hdi
```

```python
world = world.merge(hdi, on='Country', how='left')
world
```

```python
world.rename(columns={'2018': 'IDH'}, inplace=True)
world
```

```python
gini = pd.read_csv('/home/otavio/ml/datasets/world-data/gini.csv')
gini.drop(['Indicator Name', 'Country Code'], axis=1, inplace=True)
new = gini.stack().groupby(level=0).last().to_frame()
gini = pd.merge(gini, new, left_index=True, right_index=True)
gini.rename(columns={'0':'gini'}, inplace=True)
gini = gini[['Country Name', 0]]
gini
```

```python
world = world.merge(gini, how='left', left_on='Country', right_on='Country Name')
```

```python
world.drop('Country Name', axis=1, inplace=True)
world.rename(columns={0:'gini'})
```

```python
world.rename(columns={0:'gini'}, inplace=True)
world
```

```python
homi = pd.read_csv('/home/otavio/ml/datasets/world-data/homicides.csv')
homi.drop(['Code'], axis=1, inplace=True)
```

```python
homi = homi.groupby(['Entity', 'Year'])['Intentional homicides (per 100,000 people) (per 100,000 people)'].sum().to_frame().groupby(level=0).last()
```

```python
world = world.merge(homi, how='left', right_on='Entity', left_on='Country')
# world.drop('Entity', axis=1, inplace=True)
```
