```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import seaborn as sns

# visualization
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/home/otavio/ml/datasets/covid/covid.csv', encoding = "ISO-8859-1")
```

```python
data.info()
```

```python
data = data[['location', 'date', 'total_deaths']]
data.to_csv('covid.csv')
```

```python
new_data = data.pivot_table('total_deaths', ['location'], 'date')
new_data = new_data.fillna(0)
new_data = new_data.reset_index()
new_data
```

```python
countries = pd.read_csv('/home/otavio/Downloads/Country_Flags.csv')
countries
```

```python
new_data = new_data.merge(countries[['Country', 'ImageURL']], how='inner', right_on='Country', left_on='location')
new_data
```

```python
from googletrans import Translator
gt = Translator()
new_data['location'] = new_data['location'].apply(lambda x: gt.translate(text=x.replace('_', ' '), src='en', dest='pt').text)
new_data
```

```python
new_data.to_csv('covid.csv')
```
