```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rn
import seaborn as sns

# visualization
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
qlt_data = pd.read_csv('/home/otavio/ml/datasets/World_Data/quality_of_life_index_by_countries_2020.csv')
freed_data = pd.read_csv('/home/otavio/ml/datasets/World_Data/economic_freedom/economic_freedom_index2019_data.csv', error_bad_lines=False)
```

```python
print(qlt_data.info())
print(qlt_data.describe())
```

```python
print(freed_data.info())
print(freed_data.describe())
```

```python
data = freed_data.merge(qlt_data, how='inner', left_on='Country Name', right_on='Country')
data
```

```python
data['Score'].values.reshape(1, -1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
x = data['Score'].values.reshape(1, -1)
y = data['Quality of Life Index'].values.reshape(1, -1)
lr.fit(x, y)
predict = lr.predict(x)
predict
```

```python
data = data.sort_values(['Score'])
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(data['Score'], data['Quality of Life Index'])
# plt.plot(data['Score'], predict[0])
plt.ylabel('Qualidade de Vida')
plt.xlabel('índice de Liberdade')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("freedomXquality.png")
plt.show()
```

```python
data = data.sort_values(['Purchasing Power Index'])
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(data['Purchasing Power Index'], data['Quality of Life Index'])
# plt.plot(data['Score'], predict[0])
plt.ylabel('Qualidade de Vida')
plt.xlabel('Poder de Compra')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("powerXquality.png")
plt.show()
```
