```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rn
import seaborn as sns

# visualization
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/home/otavio/ml/datasets/world-data/food/live-stock.csv')
```

```python
data.drop('Domain', axis=1, inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data['Value'] = data['Value'].apply(lambda x: int(x))
data
```

```python
cattle = data[data['Item'] == 'Cattle']
cattle.sort_values('Value', ascending=False, inplace=True)
cattle
```

```python
cou = pd.read_csv('/home/otavio/ml/datasets/utils/countries.csv')
cou
```

```python
cattle = cattle.merge(cou, how='left', left_on='Area', right_on='Name')
```

```python
cattle.drop(['Name', 'Code', 'Item'], axis=1, inplace=True)
```

```python
cattle = cattle[cattle['Area'] != 'China, mainland']
```

```python
cattle.head(10).to_csv('cattle.csv')
```

```python
chick = data[data['Item'] == 'Chickens']
chick.sort_values('Value', ascending=False, inplace=True)
chick = chick.merge(cou, how='left', left_on='Area', right_on='Name')
chick = chick[chick['Area'] != 'China, mainland']
chick.head(10).to_csv('chicken.csv', index=False)
```

```python
pork = data[data['Item'] == 'Pigs']
pork.sort_values('Value', ascending=False, inplace=True)
pork = pork.merge(cou, how='left', left_on='Area', right_on='Name')
pork = pork[pork['Area'] != 'China, mainland']
pork.head(10).to_csv('pork.csv', index=False)
```

```python
sheep = data[data['Item'] == 'Sheep']
sheep.sort_values('Value', ascending=False, inplace=True)
sheep = sheep.merge(cou, how='left', left_on='Area', right_on='Name')
sheep = sheep[sheep['Area'] != 'China, mainland']
sheep.head(10).to_csv('sheep.csv', index=False)
```
