```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rn
import seaborn as sns

# visualization
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cons = pd.read_csv('/home/otavio/ml/datasets/alcohol-consumption/total-alcohol-consumption-per-capita-litres-of-pure-alcohol.csv')
non = pd.read_csv('/home/otavio/ml/datasets/alcohol-consumption/share-of-population-who-never-drink-alcohol.csv')
```

```python
cons.rename(columns={'Alcohol consumption (litres per capita) (liters of pure alcohol, projected estimates, 15+ years of age)': 'Rate'}, inplace=True)
```

```python
non.rename(columns={'Indicator:Alcohol, abstainers lifetime (%) - Sex:Both sexes (%)': 'Abstinence'}, inplace=True)
```

```python
countries = pd.read_csv('/home/otavio/ml/datasets/utils/countries.csv')
```

```python
cons = cons.merge(countries[['Name', 'PT', 'ImageURL']], how='left', left_on='Entity', right_on='Name')
```

```python
non = non.merge(countries[['Name', 'PT', 'ImageURL']], how='left', left_on='Entity', right_on='Name')
```

```python
cons.sort_values('Rate', ascending=False, inplace=True)
cons = cons.reset_index()
cons.index = cons.index + 1
cons.drop('index', axis=1, inplace=True)
```

```python
rank = cons.head(9)
```

```python
brazil = cons[cons['PT'] == 'Brasil']
brazil

res = pd.concat([rank, brazil])
```

```python
res['Label'] = res['PT'].astype(str) + ' (' + res.index.astype(str) + 'º)'
res
```

```python
res.to_csv('out.csv', index=False)
```
